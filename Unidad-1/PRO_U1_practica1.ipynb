## Ejemplo 1: Registro de calificaciones de 12 alumnos <a id='id1'> </a>

1. En este ejemplo generamos una lista que contiene las calificaciones de un grupo de 12 alumnos.
2. Con esa lista de valores se pretende determinar las **Medidas de tendencia central, es decir, la media, la mediana y la moda**


```python
# Lista de calificaciones del grupo a de probabilidad y estadistica

grupoa = [70, 70, 65, 90, 100, 95, 45, 80 , 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato dela variable
type(grupoa)
```

### Determinar la media del ejemplo 1 <a id='id11'> </a>

La media, también conocida como promedio, es el valor que se obtiene al dividir la suma de un conjunto de números entre la cantidad de ellos.

Para obtener la Media de un conjunto solo tienes que seguir estos sencillos pasos:

1. Determina el conjunto de valores que buscas promediar.
2. Suma los valores para obtener el total
3. Haz el conteo de la cantidad de valores en el conjunto.
4. Divide la suma del conjunto entre la cantidad de números.

**Media aritmética de la población**

$$
\mu = \sum_{i=1}^{N}=\frac{x_i}{N}=\frac{x_1+x_2+x_3+ ... x_{12}}{12}
$$

```python
# Sacar la media de las calificaciones con una formula

promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar
# el promedio

promedio2 = mean(grupoa)
print(promedio2)
```

### Determinar la mediana del ejemplo 1  <a id='id12'> </a>

La mediana es el valor que **ocupa el lugar central de todos los datos cuando éstos están ordenados** de menor a mayor 

Los pasos para sacar la mediana son:

1. **Ordena** todos los números del más pequeño al más grande.
2. Encuentra el **número del medio** del conjunto de datos.
- Si tienes una cantidad impar: Tacha el número al final de la izquierda, después el primero a la derecha, y repite el proceso hasta quedarte con un número, que será la mediana. 
- Si tienes una cantidad par, al final quedarás con dos números en el centro. Súmalos y divídelos entre 2 para obtener la mediana.


```python
# Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print(mediana)
```

### Determinar la moda del ejemplo 1   <a id='id13'> </a>

1. La moda es el valor que más se repite dentro de un conjunto de datos. 
2. A diferencia de la media y la mediana, la moda no requiere valores numéricos y puede utilizarse con datos categóricos o discretos. 

```python
# Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

## Ejemplo 2. Tiempos de 100 metros de un grupo de 12 alumnos <a id='id2'> </a>

1. En este ejemplo generamos una lista de un grupo de 10 alumnos que corren los 100 metros y se registran los tiempos de cada uno
2. Con esa lista de valores se pretende determinar las **Medidas de tendencia central, es decir, la media, la mediana y la moda**


```python
# Definimos una varaiable llamada tiempos100mts de tipo lista, que contiene los
# tiempos que tardaron los 10 alumnos para correr los 100 metros
tiempos100mts = [ 15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.15, 20.65, 15.10, 17.20]

# Desplegamos los valores de la lista
print(tiempos100mts)
```

### Funcion mean para calcular el promedio

```python
# Usamos la funcion mean para determinar el tiempo promedio de la lista de valores

print("El tiempo promedio de los corredores es=>", mean(tiempos100mts) )
```

### Funcion median para calcular la mediana

```python
# Usamos la funcion median  para determinar el tiempo promedio de la lista de valores

# definimos una variable llammada mediana par calcular la mediana de la lista con la funcion
# median
mediana = median(tiempos100mts)
print("La media de los tiempos de los corredores es=>", mediana )
```

### Funcion mode para calcular la moda

```python
# definimos una variable llammada moda par calcular la moda  de la lista con la funcion
# mode

moda = mode(tiempos100mts)
print("La moda de los tiempos es=>", moda)
```

### Función multimode para calcular múltiples modas (por ejemplo, en caso de que sea bimodal)

```python
multimodas = multimode(tiempos100mts)
print("La moda de los tiempos es=>", multimodas)
```

## Ejemplo 3: Histograma de 2 grupos <a id='id3'> </a>

1. En este ejemplo generamos una lista aleatoria con 100 numeros reales que representa una Muestra de la producción diaria de 100 telares para alfombras (medidas en metros) para la fabrica 1
2. Generamos otra lista de la fabrica 2 con la producción diaria de 100 telares para alfombras
2. Con esa lista de valores se pretende determinar las **Medidas de tendencia central, es decir, la media, la mediana y la moda**
3. Graficar un Histograma para cada fabrica

```python
# librería para generar números aleatorios
import random
```

```python
# basado en el LIBRO estadistica para administracion pagina 12
# creamos una lista vacia  que representa una Muestra de la producción diaria de
# 100 telares para alfombra (en metros)

fabrica1 = []

# Hacemos un ciclo para generar 100  numeros aleatorios
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  # numero_aleatorio
  # la funcion round la usamos para redondear a 2 numeros decimales
  numero_aleatorio = round( random.uniform( 50, 100) , 2)
  # Agregamos el numero aleatorio a la lista de valores
  fabrica1.append(numero_aleatorio)

print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round( random.uniform( 50, 100) , 2)
  fabrica2.append(numero_aleatorio)

print(fabrica2)
```

```python
# Importamos la libreria matplotlib para crear visualizaciones de datos
import matplotlib.pyplot as plt
```

```python

plt.hist(fabrica1)
plt.title("producción diaria de la Fabrica 1")
```

```python

plt.hist(fabrica2)
plt.title("producción diaria de la Fabrica 2")
```

## Tablas de frecuencias


```python
# basado en el LIBRO estadistica para administracion pagina 12
# creamos una lista vacia  que representa la edad de 100 personas

edades = []

# Hacemos un ciclo para generar 100  numeros aleatorios qe representan la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  # numero_aleatorio
  numero_aleatorio = random.randint( 18, 70)
  # Agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los números aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df
```

```python
# Calcular la tabla de distribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
print(tabla_frecuencias)
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

## Regla de sturges

1. Es una regla que sirve para calcular el número de clases o intervalos idóneo en los que se debe dividir un conjunto de datos.
2. Establece que **el número de clases es igual a uno más el logaritmo en base dos del número total de datos**.

$
c = 1 + log _2 (N)
$

donde $N$  es el tamaño de la muestra.

```python
# Importamos la librería math de funciones matemáticas 

import math
```

```python
# Volvemos a usar la lista edades 

print(edades)
```

```python
# Crear un DataFrame con las edades 
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el número de intervalos utilizando la regla de Sturges

n = len(edades)

K = int( 1 + math.log2(100))
print("Valor de n = ", n , " \nNúmero de intervalos =", K)
```

```python
# Crear la tabla de frecuencia

tabla_frecuencia = pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python
plt.hist(edades, alpha = 0.5)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores estandar")
```

```python
# Crear un histograma basado en la tabla de frecuencia

plt.hist(df['Edad'], bins=K, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de Sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```
