### Determinar la media del ejemplo 1 <a id='id11'> </a>

La media, tambin conocida como promedio, es el valor que se obtiene al dividir la suma de un conjunto de nmeros entre la cantidad de ellos.

Para obtener la Media de un conjunto solo tienes que seguir estos sencillos pasos:

1. Determina el conjunto de valores que buscas promediar.
2. Suma los valores para obtener el total
3. Haz el conteo de la cantidad de valores en el conjunto.
4. Divide la suma del conjunto entre la cantidad de nmeros.

**Media aritmtica de la poblacin**

$$
\mu = \sum_{i=1}^{N}=\frac{x_i}{N}=\frac{x_1+x_2+x_3+ ... x_{12}}{12}
$$

```python
# Sacar la media de las calificaciones con una formula

promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar
# el promedio

promedio2 = mean(grupoa)
print(promedio2)
```

### Determinar la mediana del ejemplo 1  <a id='id12'> </a>

La mediana es el valor que **ocupa el lugar central de todos los datos cuando stos estn ordenados** de menor a mayor 

Los pasos para sacar la mediana son:

1. **Ordena** todos los nmeros del ms pequeo al ms grande.
2. Encuentra el **nmero del medio** del conjunto de datos.
- Si tienes una cantidad impar: Tacha el nmero al final de la izquierda, despus el primero a la derecha, y repite el proceso hasta quedarte con un nmero, que ser la mediana. 
- Si tienes una cantidad par, al final quedars con dos nmeros en el centro. Smalos y divdelos entre 2 para obtener la mediana.


```python
# Determinar el valor de la mediana con la funcion median

mediana = median(grupoa)
print(mediana)
```

### Determinar la moda del ejemplo 1   <a id='id13'> </a>

1. La moda es el valor que ms se repite dentro de un conjunto de datos. 
2. A diferencia de la media y la mediana, la moda no requiere valores numricos y puede utilizarse con datos categricos o discretos. 

```python
# Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

## Ejemplo 2. Tiempos de 100 metros de un grupo de 12 alumnos <a id='id2'> </a>

### Funcion mode para calcular la moda

```python
# definimos una variable llammada moda par calcular la moda  de la lista con la funcion
# mode

moda = mode(tiempos100mts)
print("La moda de los tiempos es=>", moda)
```

## Tablas de frecuencias  <a id="id7"> </a>


```python
# basado en el LIBRO estadistica para administracion pagina 12
# creamos una lista vacia  que representa la edad de 100 personas

edades = []

# Hacemos un ciclo para generar 100  numeros aleatorios qe representan la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  # numero_aleatorio
  numero_aleatorio = random.randint( 18, 70)
  # Agregamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
# Crear un DataFrame de pandas con los nmeros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df
```

```python
# Calcular la tabla de distribucin de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de distribucin de frecuencias
print("Tabla de Distribucin de Frecuencias:")
print(tabla_frecuencias)
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

## Regla de sturges  <a id="id8"> </a>

1. Es una regla que sirve para calcular el nmero de clases o intervalos idneo en los que se debe dividir un conjunto de datos.
2. Establece que **el nmero de clases es igual a uno ms el logaritmo en base dos del nmero total de datos**.

$
c = 1 + log _2 (N)
$

donde $N$  es el tamao de la muestra.

```python
# Importamos la librera math de funciones matemticas 

import math
```

```python
# Volvemos a usar la lista edades 

print(edades)
```

```python
# Crear un DataFrame con las edades 
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el nmero de intervalos utilizando la regla de Sturges

n = len(edades)

K = int( 1 + math.log2(100))
print("Valor de n = ", n , " \nNmero de intervalos =", K)
```

```python
# Crear la tabla de frecuencia

tabla_frecuencia = pd.cut( df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python
plt.hist(edades, alpha = 0.5)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores estandar")
```

```python
# Crear un histograma basado en la tabla de frecuencia

plt.hist(df['Edad'], bins=K, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de Sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```

### 9. Diagrama de Cajas y bigotes <a id="id9"> </a>

1. Las lneas se usan para indicar la variabilidad fuera de los cuartiles superior e inferior. 
2. Los valores atpicos se representan a veces como puntos individuales que estn en lnea con los bigotes. 
3.  **Si los datos son simtricos, la mediana estar en el centro de la caja**. 
4. Si los datos estn **sesgados**, la mediana estar ms cerca de la parte superior o inferior de la caja.
5. Los extremos de arriba y abajo de la caja indican los **cuantiles, o percentiles, 25 y 75**. Estos dos cuantiles tambin se conocen como cuartiles, porque separan cuartos (25 %) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartlico (IQR)
6. Los diagramas caja-bigotes son muy tiles para comparar dos o ms conjuntos de datos.
